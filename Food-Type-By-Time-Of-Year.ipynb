```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
search_values_df = pd.read_csv('search-values.csv')
```

```python
search_values_df
```

```python
new_df = search_values_df.set_index('id')
```

```python
week_numbers = []
for date in new_df['week_id']:
    numbers = date[5:7]
    week_numbers.append(numbers)
    
new_df['number_of_week'] = week_numbers
```

```python
new_df.head()
```

```python
grouped = new_df.groupby(['id','number_of_week']).mean()
```

```python
grouped
```

```python
apple_products = search_values_df[(search_values_df.id.isin(['apple','apple-au','apple-cider','apple-pie']))]
```

```python
week_num = []
for date in apple_products['week_id']:
    numbers = date[5:7]
    week_num.append(numbers)
    
apple_products['number_of_week'] = week_num
```

```python
apple_products = apple_products.groupby(['id','number_of_week']).mean()
```

```python
apple_products.head()
```

```python
apple_products.unstack(level=0).plot(kind='line')
```
