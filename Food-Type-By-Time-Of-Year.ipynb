```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
search_values_df = pd.read_csv('search-values.csv')
```

```python
search_values_df.id.unique()
```

```python
week_numbers = []
for date in search_values_df['week_id']:
    numbers = date[5:7]
    week_numbers.append(numbers)
    
search_values_df['number_of_week'] = week_numbers
```

```python
new_df = search_values_df.set_index('id')
```

```python
grouped = new_df.groupby(['id','number_of_week']).mean()
```

```python
grouped
```

The DataFrame, grouped, averages the indexed interest value per week for each food. 

```python
apple_products = search_values_df[(search_values_df.id.isin(['apple','apple-au','apple-cider','apple-pie']))]
```

```python
apple_products = apple_products.groupby(['id','number_of_week']).mean()
```

```python
apple_products.head()
```

```python
apple_products.unstack(level=0).plot(kind='line')
plt.xlabel('Week')
plt.ylabel('Search Interest Level')
plt.title('Average Interest Through the Year of Apple Products')
```

```python
wine_preferences = search_values_df[(search_values_df.id.isin(['red-wine',
       'white-wine']))]
```

```python
wine_preferences = wine_preferences.groupby(['id','number_of_week']).mean()
```

```python
wine_preferences.unstack(level=0).plot(kind='line')
plt.xlabel('Week')
plt.ylabel('Search Interest Level')
plt.title('Average Interest Through the Year of Red-Wine vs White-Wine')
```
