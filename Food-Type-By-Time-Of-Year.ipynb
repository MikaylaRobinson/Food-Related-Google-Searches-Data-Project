```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
search_values_df = pd.read_csv('search-values.csv')
```

```python
search_values_df
```

```python
new_df = search_values_df.set_index('id')
```

```python
week_numbers = []
for date in new_df['week_id']:
    numbers = date[5:7]
    week_numbers.append(numbers)
    
new_df['number_of_week'] = week_numbers
```

```python
new_df.head()
```

```python
grouped = new_df.groupby(['id','number_of_week']).mean()
```
