```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
search_values_df = pd.read_csv('search-values.csv')
```

```python
spinach_df = search_values_df.loc[search_values_df['id'].isin(['spinach'])]
```

```python
peak_interest = spinach_df.loc[search_values_df['value'].isin(['100'])]
```

```python
peak_interest
```

Seeing that the height of interest was not during the 2006 outbreak, I will look at just 2006. 

```python
year = []
for date in spinach_df['week_id']:
    numbers = date[0:4]
    year.append(numbers)

spinach_df['year'] = year
```

```python
interest_during_outbreak = spinach_df.loc[spinach_df['year'].isin(['2006'])]
```

```python
interest_during_outbreak
```
